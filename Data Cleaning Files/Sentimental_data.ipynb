```python
full_data =  [data_1, data_2, data_3, data_4, data_5,data_6]
```

```python
import pandas as pd

# Assuming you have the JSON data stored in a variable called 'data'

# Define an empty list to store the extracted article data
article_data = []
for data in full_data :
# Iterate over the articles and extract the required fields
 for article in data['results']:
    article_id = article['id']
    publisher_name = article['publisher']['name']
    title = article['title']
    author = article['author']
    published_utc = article['published_utc']
    article_url = article['article_url']
    tickers = ', '.join(article.get('tickers', []))
    image_url = article.get('image_url', '')
    description = article.get('description', '')
    
    # Create a dictionary with the extracted article data
    article_dict = {
        'id': article_id,
        'publisher': publisher_name,
        'title': title,
        'author': author,
        'published_utc': published_utc,
        'article_url': article_url,
        'tickers': tickers,
        'image_url': image_url,
        'description': description
    }
    
    # Append the article dictionary to the list
    article_data.append(article_dict)

# Create a DataFrame from the list of article dictionaries
article_df = pd.DataFrame(article_data)

# Print the resulting DataFrame
(article_df)
```

```python
 import nltk
 nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd

# Assuming you have the DataFrame stored in a variable called 'article_df'

# Initialize sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Create empty lists to store the sentiment scores
neg_scores = []
neu_scores = []
pos_scores = []
compound_scores = []

# Iterate over the rows of the DataFrame and calculate sentiment scores
for index, row in article_df.iterrows():
    text = row['description']  # Adjust the column name according to your DataFrame
    
    # Calculate sentiment scores using VADER
    sentiment_scores = sid.polarity_scores(text)
    
    # Append the scores to the respective lists
    neg_scores.append(sentiment_scores['neg'])
    neu_scores.append(sentiment_scores['neu'])
    pos_scores.append(sentiment_scores['pos'])
    compound_scores.append(sentiment_scores['compound'])

# Add the sentiment scores to the DataFrame as new columns
article_df['neg'] = neg_scores
article_df['neu'] = neu_scores
article_df['pos'] = pos_scores
article_df['compound'] = compound_scores

# Print the updated DataFrame
print(article_df)
```

```python
csv_file_path = 'article_data.csv'

# Convert the DataFrame to CSV
article_df.to_csv(csv_file_path, index=False)
```
